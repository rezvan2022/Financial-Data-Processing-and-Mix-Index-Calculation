```python
# Define the Mix_Index_ratio with 1 decimal place rounding
Mix_Index_ratio = [round(k * 0.2, 1) for k in range(6)]

# Loop through each ratio and create corresponding columns in KPI_data
for ratio in Mix_Index_ratio:
    column_name = f"Mix Index {ratio:.1f} (LTM-FM) + FM"
    KPI_data[column_name] = None  # Initialize with None to handle missing data

# Loop through the data and calculate values for each ratio
for i in range(len(KPI_data)):
    for k, ratio in enumerate(Mix_Index_ratio):
        column_name = f"Mix Index {ratio:.1f} (LTM-FM) + FM"

        # Check if both 'LTM Index' and 'Final Month' are valid values
        ltm_value = KPI_data.at[i, 'LTM Index']
        final_month_value = KPI_data.at[i, 'Final Month']
        if pd.notna(ltm_value) and ltm_value != "NA" and pd.notna(final_month_value) and final_month_value != "NA":
            KPI_data.at[i, column_name] = final_month_value + ratio * (ltm_value - final_month_value)
        else:
            KPI_data.at[i, column_name] = "NA"
```
